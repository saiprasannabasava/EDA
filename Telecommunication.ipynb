# Univariate analysis(numerical columns)

```python

def plot_box(df):
    for column in df.columns:
        if df[column].dtype == np.int64 or df[column].dtype == np.float64:
            if column == 'SeniorCitizen':
                df['SeniorCitizen'].value_counts(normalize=True).plot(kind='box')
                plt.title('SeniorCitizen')
                plt.show()
            else:
                plot_box = df.boxplot( column=column)
                plt.title(f"{column}")
                plt.show()
plot_box(df)
```


* Box plot is used for univariate and bivariate analysis.
* Box plot is used to find the outliers.
* Here we draw box plot for numerical columns and we didn't find any outliers in it.

# Univariate analysis(Categorical column)

```python
def plot_bar(df):
    for column in df.columns:
        if df[column].dtype == np.object:
            if column == 'customerID':
                print('No need to plot for "customerID" column.')
            else:
                counts = df[column].value_counts()
                counts.plot(kind='bar')
                plt.title(column)
                plt.show()
plot_bar(df)
```

* From above countplot dependents customers are less compared to 
  non dependents.
* From above countplot phoneservice are high.
* From above countplot most of the customers has no multiplelines 
  connection.
* From above countplot more customers uses fiber optic 
  internetService compared to others.
* Most customres has no onlineSecurity.
* most of the customers has no OnlineBackup
* Most of the customers has no deviceprotection

# Bivariate analysis(cat vs num)

```python
def plot_box(df):
    a = df["Churn"]
    for column in df.columns:
        if df[column].dtype == np.int64 or df[column].dtype == np.float64 :
            plot_box = df.boxplot( by="Churn",column=column)
            plt.title(f"Churn vs {column}")
            plt.show()
plot_box(df)
```

# Bivariate analysis(cat vs cat)

```python
df.info()
```

```python
def plot_stack(df):
    a = df["Churn"]
    for column in df.columns:
        if df[column].dtype == np.object:
            if column=='customerID':
                print('No need to plot for "customerID" column.')
            else:    
                plot_stck = pd.crosstab(index=df[column], columns=a)
                plot_stck.plot(kind='bar', stacked=True)
                plt.title(f" Churn vs {column}")
                
                plt.show()
plot_stack(df)
```

# Conclusions:
 * Customers subscribed to specific services, such as online security, 
   online backup, device protection, and tech support, may be more likely
   to churn. 
*  The telecom company should investigate the quality, availability, and 
   value proposition of these services to improve customer retention.
* The contract type, payment method, and paperless billing options may 
  influence customer churn.
* Consider offering flexible contract options, convenient payment 
  methods, and incentives for paperless billing to encourage customer 
  loyalty.
* High monthly charges and total charges may contribute to customer 
  churn. Evaluate pricing strategies and ensure that customers perceive 
  the value of the services they receive.
* Demographic factors may play a role in churn. 
* Tailor marketing campaigns and customer retention strategies based on 
  gender, age range, and family status

# Recommendations
 Conduct customer surveys or interviews to gain qualitative insights 
into the reasons behind churn and gather feedback on service quality 
and customer experience.
* Implement proactive customer retention measures, such as 
  personalized offers, loyalty programs, or customer service 
  enhancements, for customers identified as high-risk for churn.
* Regularly review and update service offerings to align with customer
  needs and preferences, considering market trends and competitors.
* Invest in improving customer support channels, ensuring prompt and 
  effective resolutions to customer issues and concerns.
* Develop data-driven churn prediction models to identify at-risk 
  customers and implement targeted interventions to reduce churn.
